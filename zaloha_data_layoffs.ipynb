```python
import tiktoken
import time
import pandas as pd
import os

# Načtení vstupního dataframu
df_final_merge = df_final_merge[df_final_merge['text'].notna()]
df_final_merge['text'] = df_final_merge['text'].astype(str)

# Zjistit, které odpovědi už máme
results = []
results_path = "je_tam_duvod_vypovedi.csv"

if os.path.exists(results_path):
    existing_results = pd.read_csv(results_path)
    processed_ids = set(existing_results['id'].tolist())
    results = existing_results.to_dict(orient='records')
    print(f"Načteno {len(results)} již zpracovaných výsledků.")
else:
    processed_ids = set()
    print("Nebyl nalezen žádný předchozí výstupní soubor. Začíná se od začátku.")

# Filtrování nezpracovaných řádků
df_to_process = df_final_merge[~df_final_merge['index'].isin(processed_ids)]

# Zpracování zbývajících článků
for index, row in df_to_process.iterrows():
    index_clanku = row['index']
    text_clanku = row['text']
    enc = tiktoken.encoding_for_model("gpt-4o")
    num_tokens = len(enc.encode(text_clanku))
    print("Tokenů:", num_tokens)
    if num_tokens < 150000:
        completion = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system",
                "content": "You are given an article that contains an information of a lay off of a company. Read the article and evaluate whether it contain a reason of the lay off. Provide answer as a single word. Return YES if the article contains the reason and NO if it does not contain the reason."},
                {"role": "user",
                "content": "here is the article:" + text_clanku}
            ]
        )

        response_text = completion.choices[0].message.content
        print(f"odpoved pro id {index_clanku}: {response_text}")
        time.sleep(65)

        results.append({
            'id': index_clanku,
            'odpoved': response_text
        })

    # PRŮBĚŽNÉ UKLÁDÁNÍ každých 5 článků
    if len(results) % 5 == 0:
        df_results = pd.DataFrame(results)
        df_results.to_csv(results_path, index=False)
        print(f"Průběžně uloženo {len(results)} výsledků do souboru.")

# Finální uložení po skončení cyklu
df_results = pd.DataFrame(results)
df_results.to_csv(results_path, index=False)
print("Finální uložení hotovo.")
```
