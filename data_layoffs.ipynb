```python
import pandas as pd
import json
import csv
from openai import OpenAI
muj_api_klic = ""
client = OpenAI(api_key=muj_api_klic)
```

```python
df1 = pd.read_json('vzorek_scraping_J.json')
df2 = pd.read_json('vzorek_scraping_l.json')
df3 = pd.read_json('vzorek_scraping_v.json')
df_carodejnice = pd.read_json('dataset_30_04.json')
df_carodejnice['url'] = df_carodejnice['url'].str.strip()
```

```python
spojeny_df = pd.concat([df1, df2, df3, df_carodejnice], axis=0, ignore_index=True)
```

```python
spojeny_df.to_json("vzorek_data_layoffs.json", orient='records', indent=4)
```

```python
df_selected = spojeny_df[['url', 'text']]
df_selected['url'] = df_selected['url'].str.strip()
```

```python
dfAll = pd.read_csv('Layoffs.fyi_Tech_Layoffs_Tracker_Original.csv')
dfAll = dfAll.rename(columns={'Source': 'url'})
dfAll.insert(0, 'index', range(0, len(dfAll)))
dfAll['url'] = dfAll['url'].str.strip()
dfAll.tail(3)
```

```python
merged_df = pd.merge(dfAll, df_selected, on='url', how='left')
merged_df = merged_df.drop_duplicates()
df_no_duplicates = merged_df.drop_duplicates(subset='index', keep='first')
df_no_duplicates
```

```python
df_no_duplicates['text'] = df_no_duplicates['text'].astype(str)
```

```python
df_no_duplicates.to_csv("ocisteny_layoffs.csv", index=False)
df_no_duplicates 
```

```python
df_nedele = pd.read_json('dataset_nedele_2.json')
df_nedele['url'] = df_nedele['url'].str.strip()
df_nedele.to_csv("nedelni_layoffs.csv", index=False)
df_nedele = df_nedele.drop_duplicates(subset="url")
```

```python
df_carodejnice.to_csv("posledni_layoffs.csv", index=False)
df_carodejnice = df_carodejnice.drop_duplicates(subset="url")
```

```python
df_merge_nedele = pd.merge(df_no_duplicates, df_nedele, on='url', how='left')
```

```python
df_final_merge = pd.merge(df_merge_nedele, df_carodejnice, on='url', how='left')
```

```python
df = pd.read_csv("posledni_layoffs.csv", encoding='utf-8') 
df_final_merge = df[df['text'].notna()]
df_final_merge['text'] = df_final_merge['text'].astype(str)

# prázdný list, kam si budu ukládat odpovědi chatu
results = []

# tady začíná for loop. df.head(3) říká, že loop proběhne pouze první tři řádky dataframu.
# Toto číslo můžete zvýšit, popř. head úplně odstranit, až budete mít odladěný prompt.
for index, row in df_merge_nedele.iterrows():

    # do proměnných si uložím id zvířete a obsah ze sloupce concatenated, kde je úplně vše z předchozích sloupců.
    # můžete si vytvořit vlastní proměnné pouze z těch sloupců, které vy konkrétně potřebujete.
    index_clanku = row['index']
    text_clanku = row['text']
    
    # printy jsou zde na prohlížení a odladění. V reálu byste je pak zakomentovaly.
    # print(index_clanku)
    # print(text_clanku)

    # tady je vám dobře známý completions :-)
    # krom změn system message a user message můžete přidat ještě assistant message na přidání příkladu
    # můžete přidat parametr temperature, max_tokens aneb response_format, máte-li vhodnou verzi pythonu
    completion = client.chat.completions.create(
                    model="gpt-4o-mini",
                    messages=[
                            {"role": "system",
                             "content": "You are given an article that contains an information of a lay off of a company. Read the article and evaluate whether it contain a reason of the lay off. Provide answer as a single word. Return YES if the article contains the reason and NO if it does not contain the reason."},
                            {"role": "user",
                             "content": "here is the article:" + text_clanku}
                            ]
                    )
    
    # zde vyextrahujete obsah odpovědi
    response_text = completion.choices[0].message.content
    
    # a odpověď si vyprintíte - můžete později zakomentovat
    print("odpoved:")
    # print(response_text)
    # print("-----------------")
    # print()

    # výsledek aktuálního běhu for loopu uložíte do listu results
    results.append({
        'id': index_clanku,
        'odpoved': response_text
    })

# PRŮBĚŽNÉ UKLÁDÁNÍ každých 5 článků - dořešit aby to ukazovalo vždy když se udělá průběžné ukládání
    if (index + 1) % 5 == 0:
        df_results = pd.DataFrame(results)
        df_results.to_csv("je_tam_duvod_vypovedi_prubezne.csv", index=False)
        print(f"Průběžně uloženo {len(results)} výsledků do souboru.")
        

# Konečné uložení po ukončení celého cyklu
df_results = pd.DataFrame(results)
df_results.to_csv("je_tam_duvod_vypovedi.csv", index=False)
# print("Finální uložení hotovo.")
```
